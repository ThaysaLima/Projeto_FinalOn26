### Descrição das colunas:
- *Ano de eleição*: Apenas informar o ano em que o candidato se candidatou, podendo ser eles 2014, 2016, 2018, 2020 ou 2022.  
- *Cargo*: Representa o cargo para o qual o candidato está concorrendo, como presidente, governador, prefeito, senador, deputado federal, deputado estadual, vereador, entre outros. 
- *Cor/raça*: Refere-se à autodeclaração da cor ou raça do candidato, comumente classificadas como branca, preta, parda, amarela ou indígena.
- *Faixa etária*: Representa a faixa etária à qual o candidato pertence, por exemplo, 18-24 anos, 25-34 anos, 35-44 anos, 45-54 anos, 55-64 anos, 65 anos ou mais.  
- *Gênero*: Indica o gênero do candidato, geralmente categorizado como masculino ou feminino.  
- *Grau de instituição*: Refere-se ao nível educacional do candidato, podendo ser ensino fundamental incompleto, ensino fundamental completo, ensino médio incompleto, ensino médio completo, ensino superior incompleto ou ensino superior completo.  
- *Região*: Indica a região geográfica do Brasil à qual a UF pertence, como Norte, Nordeste, Centro-Oeste, Sudeste ou Sul.  
- *Situação de candidatura*: Informar se o candidato esta apto ou inapto.
- *Situação de totalização*: Se o candidato foi eleito ou não.
- *Turno*: Informa o turno da eleição em que o candidato está concorrendo, podendo ser primeiro turno ou segundo turno. 
- *UF*: Refere-se à unidade federativa na qual o candidato está concorrendo, como São Paulo (SP), Rio de Janeiro (RJ), Minas Gerais (MG), etc.  
- *Quantidade de candidatos*: Refere-se ao número total de candidatos que participaram da eleição para a combinação de características especificadas.  
- *Quantidade de candidatos eleitos*: Indica quantos dos candidatos obtiveram sucesso e foram eleitos.  
- *Quantidade de candidatos para o 2º turno*: Mostra quantos candidatos foram para o segundo turno em eleições que possuem essa etapa.  
- *Quantidade de candidatos não eleitos*: Representa o número de candidatos que não foram eleitos ao final da eleição.  


```python
df_candidatos = pd.read_csv('../datasets/dataset_final.csv', encoding='latin1', sep=';', na_values=['#NULO', '#NE'])
df_candidatos
```

```python
#Irei excluir 2 colunas que não irão ser necessárias para a aálise.
excluir_colunas = ['Detalhe da situação de candidatura', 'Data de carga']
df_candidatos = df_candidatos.drop(excluir_colunas, axis=1)
```

```python
df_candidatos.info()
```

```python
#Irei mudar algumas colunas que estão tipo object, pois estão separadas por virgula. 
colunas_int = ['Quantidade de candidatos', 'Quantidade de candidatos eleitos', 'Quantidade de candidatos para o 2º turno', 'Quantidade de candidatos não eleitos','Quantidade de candidatos suplentes', 'Quantidade de candidatos não informados']

df_candidatos[colunas_int] = df_candidatos[colunas_int].replace({',': '.'}, regex=True).astype(float).astype(int)
df_candidatos.info()
```

```python
df_candidatos.describe()
```

```python
#Irei aqui mostrar quantoas cargos existem nesse dataset.
resultados = df_candidatos.groupby(['Cargo', 'Ano de eleição']).size()
resultados
```

```python
#Número de candidatas, mulheres, eleitas no segundo turno.
eleitas_segundo_turno = df_candidatos[df_candidatos['Quantidade de candidatos para o 2º turno'] == True]
contagem_por_ano_genero = eleitas_segundo_turno.groupby(['Ano de eleição', 'Gênero']).size().unstack()

with sns.color_palette("Paired"):
    ax = contagem_por_ano_genero.plot(kind='bar', stacked=True)

for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height}', (x + width/2, y + height), ha='center', va='bottom')

plt.ylabel('Número de Candidatos Eleitos no Segundo Turno')
plt.title('Número de Candidatas Eleitas no Segundo Turno por Ano')
plt.legend(title='Gênero', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
#Número de candidatos eleitos e não eleitos por ano.
df_filtrado = df_candidatos[df_candidatos['Situação de totalização'].isin(['Eleito', 'Não eleito'])]

contagem_por_ano_situacao = df_filtrado.groupby(['Ano de eleição', 'Situação de totalização']).size().unstack()

with sns.color_palette("hls", 14):
    ax = contagem_por_ano_situacao.plot(kind='bar', stacked=True)

for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height}', (x + width/2, y + height), ha='center', va='bottom')

plt.ylabel('Número de Candidatos')
plt.title('Número de Candidatos Eleitos e Não Eleitos por Ano')
plt.legend(title='Situação', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
#número de candidatos por gênero e ano
df_filtrado = df_candidatos[df_candidatos['Gênero'] != 'Não divulgável']

contagem_por_ano_genero = df_filtrado.groupby(['Ano de eleição', 'Gênero']).size().unstack()

with sns.color_palette("husl"):
    ax = contagem_por_ano_genero.plot(kind='bar', stacked=True)

for p in ax.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height}', (x + width/2, y + height), ha='center', va='bottom')

plt.ylabel('Número de Candidatos')
plt.title('Número de Candidatos por Gênero e Ano (Excluindo "Não divulgável")')
plt.legend(title='Gênero', loc='upper left', bbox_to_anchor=(1, 1))
plt.show()
```

```python
#distribuiçao de homens e mulheres, por cor/raça
df_mulheres = df_candidatos[df_candidatos['Gênero'] == 'Feminino']

df_homens = df_candidatos[df_candidatos['Gênero'] == 'Masculino']

plt.figure(figsize=(15, 6))

plt.subplot(1, 2, 1)
ax1 = sns.countplot(x='Cor/raça', data=df_mulheres)
plt.title('Distribuição por Cor/Raça - Mulheres')
plt.xlabel('Cor/raça')
plt.xticks(rotation=45, ha='right')

for p in ax1.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax1.annotate(f'{height}', (x + width/2, y + height), ha='center', va='bottom')

plt.subplot(1, 2, 2)
ax2 = sns.countplot(x='Cor/raça', data=df_homens)
plt.title('Distribuição por Cor/Raça - Homens')
plt.xlabel('Cor/raça')
plt.xticks(rotation=45, ha='right')

for p in ax2.patches:
    width = p.get_width()
    height = p.get_height()
    x, y = p.get_xy() 
    ax2.annotate(f'{height}', (x + width/2, y + height), ha='center', va='bottom')  

plt.tight_layout()
plt.show()
```
