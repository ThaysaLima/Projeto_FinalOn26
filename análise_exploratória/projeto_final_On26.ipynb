## Mulheres em cargos gorvernamentais no Brasil.

Nesse dataset irei analisar, em comparativos aos homens, mulheres nos cargos governamentais. Nos anos de 2014, 2016, 2018, 2020 e 2022.
Os dados foram coletados atráves do site TSE (Tribuna Superior Eleitoral). 
O site incialmente me deu um arquivo csv com mais de 70 colunas, mas pesquisando mais um pouco, o próprio site do TSE permitia o download filtrado das informações desejadas.

```python
#importando bibliotecas,
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#primeiro de tudo, irei juntar os 5 datasets, relativos aos anos de 2014, 2016, 2018, 2020 e 2022. 
diretorio = '/Users/thaysalima/Desktop/RePrograma./Python Fundamentos e análise de dados/Projeto_Final/Projeto_FinalOn26/datasets/CandidatosBrasil'

dataframes = []

for arquivo in os.listdir(diretorio):
    if arquivo.endswith(".csv"):
        caminho_arquivo = os.path.join(diretorio, arquivo)

        df = pd.read_csv(caminho_arquivo, encoding='latin1', sep=';', na_values=['#NULO', '#NE'])

        df = df.fillna({'campo_numerico': -1})

        dataframes.append(df)

df_final = pd.concat(dataframes, ignore_index=True)

df_final.to_csv('/Users/thaysalima/Desktop/RePrograma./Python Fundamentos e análise de dados/Projeto_Final/Projeto_FinalOn26/datasets/dataset_final.csv', encoding='latin1', sep=';', index=False)
```

### Descrição das colunas:
- *Ano de eleição*: Apenas informar o ano em que o candidato se candidatou, podendo ser eles 2014, 2016, 2018, 2020 ou 2022.  
- *Cor/raça*: Refere-se à autodeclaração da cor ou raça do candidato, comumente classificadas como branca, preta, parda, amarela ou indígena.  
- *Faixa etária*: Representa a faixa etária à qual o candidato pertence, por exemplo, 18-24 anos, 25-34 anos, 35-44 anos, 45-54 anos, 55-64 anos, 65 anos ou mais.  
- *Gênero*: Indica o gênero do candidato, geralmente categorizado como masculino ou feminino.  
- *Grau de instituição*: Refere-se ao nível educacional do candidato, podendo ser ensino fundamental incompleto, ensino fundamental completo, ensino médio incompleto, ensino médio completo, ensino superior incompleto ou ensino superior completo.  
- *Município*: Indica o município onde o candidato está concorrendo à eleição.  
- *Cargo*: Representa o cargo para o qual o candidato está concorrendo, como presidente, governador, prefeito, senador, deputado federal, deputado estadual, vereador, entre outros.  
- *UF*: Refere-se à unidade federativa na qual o candidato está concorrendo, como São Paulo (SP), Rio de Janeiro (RJ), Minas Gerais (MG), etc.  
- *Região*: Indica a região geográfica do Brasil à qual a UF pertence, como Norte, Nordeste, Centro-Oeste, Sudeste ou Sul.  
- *Turno*: Informa o turno da eleição em que o candidato está concorrendo, podendo ser primeiro turno ou segundo turno.  
- *Quantidade de candidatos*: Refere-se ao número total de candidatos que participaram da eleição para a combinação de características especificadas.  
- *Quantidade de candidatos eleitos*: Indica quantos dos candidatos obtiveram sucesso e foram eleitos.  
- *Quantidade de candidatos para o 2º turno*: Mostra quantos candidatos foram para o segundo turno em eleições que possuem essa etapa.  
- *Quantidade de candidatos não eleitos*: Representa o número de candidatos que não foram eleitos ao final da eleição.  


```python
df_candidatos_br = pd.read_csv('../datasets/dataset_final.csv', encoding='latin1', sep=';', na_values=['#NULO', '#NE'])
df_candidatos_br
```

```python
df_candidatos_br.info()
```

```python
#tive que mudar o separador que antes estava por ',' e acusava ser object. Para poder mudar para o tipo int. 
colunas_int = ['Quantidade de candidatos', 'Quantidade de candidatos eleitos', 'Quantidade de candidatos para o 2º turno', 'Quantidade de candidatos não eleitos']

df_candidatos_br[colunas_int] = df_candidatos_br[colunas_int].replace({',': '.'}, regex=True).astype(float).astype(int)
```

```python
#Gráfico gênero ao longo dos anos.
contagem_genero = df_candidatos_br.groupby(['Ano de eleição', 'Gênero'])['Quantidade de candidatos'].sum().unstack()
plt.figure(figsize=(25, 8))
contagem_genero.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Distribuição de Candidatos por Gênero ao Longo dos Anos')
plt.xlabel('Ano de Eleição')
plt.ylabel('Quantidade de Candidatos')
plt.legend(title='Gênero')
plt.show()
```

```python
cargos_unicos = df_candidatos_br['Cargo'].unique()
cargos_unicos
```

```python
#Gráfico distribuição de gêneros nos cargos.
contagem_cargos = df_candidatos_br.pivot_table(index='Cargo', columns='Gênero', values='Quantidade de candidatos eleitos', aggfunc='sum', fill_value=0)

ax = contagem_cargos.plot(kind='bar', stacked=True, colormap='viridis', figsize=(12, 6))
plt.title('Distribuição de Gênero em Cargos')
plt.xlabel('Cargo')
plt.ylabel('Quantidade de Candidatos Eleitos')
plt.legend(title='Gênero', loc='upper right')
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
#gráfico por cor/raça
plt.figure(figsize=(12, 6))
sns.countplot(x='Cor/raça', hue='Gênero', data=df_candidatos_br)
plt.title('Distribuição por Cor/Raça')
plt.xlabel('Cor/raça - Todos os Anos')
plt.xticks(rotation=45, ha='right')

plt.show()
```

```python
#Gráfico por faixa etária
plt.figure(figsize=(15, 6))
sns.histplot(x='Faixa etária', hue='Gênero', data=df_candidatos_br, multiple='stack', binwidth=1)
plt.title('Distribuição por Faixa Etária')
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Insights
 Ao longo dos anos, a distribuição de candidatos por gênero mostra que as mulheres tiveram uma tendência de crescimento relativamente pequena comparada aos homens.  

 A análise envolve uma variedade de cargos, incluindo Deputado Estadual, Governador, Senador e outros. Esta diversidade proporciona uma visão abrangente da participação dos candidatos em diferentes posições políticas.  

 A distribuição de gênero em cargos políticos revela que as mulheres ainda não ocupam uma grande porcentagem de poder, apenas em cargos como prefeita e vereadora podemos ver uma pequena parcela de mulheres.  

 A distribuição por faixa etária oferece uma visão sobre a participação de diferentes faixas etárias na política. Observa-se que as poucas mulheres que estão em poderes ministrais ficam ali na faixa de 35 a 50 anos, difente dos homens que já começa na política já com seus 19/20 anos até a idade mais velha.
